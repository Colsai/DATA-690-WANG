### 10b. Making a specific data frame

```
# Let's look for them all at once. We know the admit rate for them is low- they're all good schools.
smlr_df[smlr_df["STABBR"] == 'MD'].sort_values(by = 'ADM_RATE')[0:20]
```

```
#Strangely, we are still missing Loyola. Let's take a look for Loyola, MD.
smlr_df[smlr_df["INSTNM"] == "Loyola"]
```

```
#No result. Let's try a different idea. Let's try str.contains
smlr_df[smlr_df["INSTNM"].str.contains("Loyola")]
```

```
#Creating a small function for finding names w/ contains
def name_finder(dataframe, column, name):
    find_it = dataframe[dataframe[column].str.contains(name)]
    return find_it

#Now if we needed to, we could search for all of them at once.
name_finder(smlr_df, "INSTNM", "Baltimore County")
```

With str.contains, we can find any of these Universities pretty easily...

```
name_finder(smlr_df, "INSTNM", "Johns Hopkins")
```

```
name_finder(smlr_df, "INSTNM", "College Park")
```

```
name_finder(smlr_df, "INSTNM", "Loyola")
```

We found them, so we can add them to a new data frame with .isin : 

```
#We found the names from above, so we can directly select them.
md_names = ['University of Maryland-Baltimore County',
            'Johns Hopkins University',
            'University of Maryland-College Park',
            'Loyola University Maryland']

#Now we can just add them into the df
md_names_df = smlr_df[smlr_df["INSTNM"].isin(md_names)]

md_names_df.sort_values("INSTNM")
```

```
#Or, we could directly take it by institution number. We have that information, too.
md_codes = [163268,163046,162928,163286]

md_codes_df = smlr_df[smlr_df["UNITID"].isin(md_codes)]

md_codes_df.sort_values("INSTNM")

#Done
```

### 10b. Last Step
- Calculate the min, max, and average in-state tution of the above four institutions.

* min = .min
* max = .max
* average = .mean()

```
# We have the data frame. Now we can do the next part.
print(f"${md_names_df['TUITIONFEE_IN'].min()} min")
print(f"${md_names_df['TUITIONFEE_IN'].max()} max")

# Average
method_1 = md_names_df['TUITIONFEE_IN'].sum() / len(md_names_df['TUITIONFEE_IN'])
method_2 = md_names_df['TUITIONFEE_IN'].mean()

print(f"${method_1} average")
```
