## Melting More Stuff...
What happens with multiple variables stored in one column?

```
ebola = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Cleaning_data_pandas/pydatadc_2018-tidy-master/pydatadc_2018-tidy-master/data/country_timeseries.csv')
```

```
#Head of the data
ebola.head()
```

Note: This is a small, curated dataset of Ebola. One of the issues for this data is that we want different columns- column for country, column for cases, column for deaths, etc. 

Columns we want: date, day, country, case or not, etc.

```
#First thing that we want to do is to melt the columns down. 
ebola_long = pd.melt(ebola, 
                     id_vars = ["Date", "Day"])
```

```
#Good, but the variable column is a problem.
ebola_long.head()
```

How do we split this? We can break the columns down by the underscores. 

```
#Let's try to break up a test string first, between "_"
'Cases_Guinea'.split("_")
```

OK- we need something like this to work on the columns. We want to break each of the columns by split. 

```
#Don't forget the accessor, it treats the column as a string
variable_split = ebola_long['variable'].str.split("_")
```

```
#Now let's look at variable_split
variable_split[0:10]
```

```
#This is a series (1 column data)
type(variable_split)
```

```
type(variable_split[0])
```

```
variable_split[0]
```

Use Type anytime there is a problem with looking at data- it can be very helpful to understand what you are seeing. 

```
#ebola_long['variable'] = [element[1] for element in ebola_long['variable']] 
```

```
#Treat this thing as a string, get the first thing out of it for ALL
variable_split.str.get(0)
```

```
#Treat this thing as a string, get the first thing out of it for ALL
variable_split.str[0] #Get Also Works
```

```
#So lets add these to the old dataframe ebola_long
ebola_long['stats'] = variable_split.str.get(0)
ebola_long['country'] = variable_split.str.get(1)
```

```
ebola_long.head()
```

Great, we got it!

This is pretty cool, we can just do this in a single line

```
#If we want to do this one go, however, we can also get a dataframe back
ebola_long[['stats', 'country']] = ebola_long['variable'].str.split('_',expand = True)
```

```
#So it worked. Now we can look
ebola_long.head()
```

## Variables in Both Rows AND Columns
- What happens whe we have variables in both rows and columns.  
- How do we know that there is data everywhere? 
- If you end up with the vast majority of the row is duplicated, variables are probably being stored in both rows and columns. 
- So variables are being stored in a row. 

```
weather = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Cleaning_data_pandas/pydatadc_2018-tidy-master/pydatadc_2018-tidy-master/data/weather.csv')
```

```
#Probably want year, month, day, and temperature max/min
weather
```

How do we pivot the values back up?
tmax and tmin should be column names and headers

```
#Weather_melt OR weather.melt
weather_melt = pd.melt(
    weather,
    id_vars = ['id','year','month','element'],
    var_name='day',
    value_name='temp'
)
```

```
#Melt the weather_df
weather_melt.head()
```

We fixed the data, but we still need to know if it's a min or max temperature.
If we wanted to do a model for year, month, and day, could we do it? We want the min or max temperature, so that we could do a model, but we have no way to specify this currently. 

With pivot, we can specify the min, mean, max, etc.

```
#Let's melt and pivot the table
weather_tidy = weather_melt.pivot_table(
    index = ['id', 'year', 'month', 'day'],
    columns = 'element', #What to pivot on
    values = 'temp', #Where the values go
)
```

```
weather_tidy.head()
```

```
#If you want a regular, flat dataframe, you can use reset_index
weather_tidy.reset_index()
```
